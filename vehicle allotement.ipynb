# Queue

```python
class VehicleQueue:
    def __init__(self):
        self.queue = []

    def add_vehicle(self, vehicle):
        self.queue.append(vehicle)

    def process_next_vehicle(self):
        if len(self.queue) == 0:
            print("No vehicles in the queue.")
            return

        next_vehicle = self.queue.pop(0)
        print(f"Processing vehicle {next_vehicle}...")

    def display_queue(self):
        if len(self.queue) == 0:
            print("No vehicles in the queue.")
            return

        print("Vehicles in the queue:")
        for vehicle in self.queue:
            print(vehicle)

    def pop_vehicle(self):
        if len(self.queue) == 0:
            print("No vehicles in the queue.")
            return

        vehicle = self.queue.pop()
        print(f"Removed vehicle {vehicle} from the queue.")

vehicle_queue = VehicleQueue()

while True:
    choice = input("\nEnter '1' to add a vehicle to the queue, '2' to process the next vehicle, '3' to display the queue, '4' to remove the last vehicle, or 'q' to quit: ")
    if choice == '1':
        vehicle = input("Enter the vehicle no. : ")
        vehicle_queue.add_vehicle(vehicle)
    elif choice == '2':
        vehicle_queue.process_next_vehicle()
    elif choice == '3':
        vehicle_queue.display_queue()
    elif choice == '4':
        vehicle_queue.pop_vehicle()
    elif choice == 'q':
        break
    else:
        print("Invalid choice. Please enter '1', '2', '3', '4', or 'q'.")
```

# Graph

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, node1, node2):
        if node1 in self.graph:
            self.graph[node1].append(node2)
        else:
            self.graph[node1] = [node2]

        if node2 in self.graph:
            self.graph[node2].append(node1)
        else:
            self.graph[node2] = [node1]

    def display_network(self):
        print("Road network:")
        for node, neighbors in self.graph.items():
            print(f"{node}: {', '.join(neighbors)}")

def main():
    road_network = Graph()

    while True:
        choice = input("\nEnter '1' to add a road, '2' to display the road network, or 'q' to quit: ")
        
        if choice == '1':
            node1 = input("Enter the starting point: ")
            node2 = input("Enter the destination point: ")
            road_network.add_edge(node1, node2)
        elif choice == '2':
            road_network.display_network()
        elif choice == 'q':
            break
        else:
            print("Invalid choice. Please enter '1', '2', or 'q'.")

if __name__ == "__main__":
    main()
```
